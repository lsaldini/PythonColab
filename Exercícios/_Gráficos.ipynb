#Matplotlib e Seaborn

```
# Bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
x = list(range(1,51)) #índices
y = np.random.normal(size=50)
print(x)
print(y)
```

```
# plot 
plt.plot(x, y,'r+-')
plt.show() #suprimir o texto gerado ao plotar o gráfico
```

```
# gráficos de dispersão
plt.scatter(x, y, color='blue');
```

```
plt.figure(figsize=(20,5)) #aumento o tamanho do gráfico.
plt.plot(x, y,'r+-')
plt.show()
```

```
# boxplot
plt.boxplot(y, showmeans=True);
```

```
y_outlier = np.concatenate((y, [4]))
plt.boxplot(y_outlier);
```

```
# extrair os outliers
fig = plt.boxplot(y_outlier)
```

```
for outlier in fig['fliers']:
    print(outlier.get_ydata())
```

```
plt.violinplot(y);
```

```
sns.violinplot(data=y, color='0.8')
sns.stripplot(data=y);
```

```
#histogramas
plt.hist(y, bins=20, color='red');
```

```
y2 = np.random.normal(size=50)
plt.hist(y, alpha=0.7)
plt.hist(y2, alpha=0.7)
plt.show()
```

```
plt.hist(y, alpha=0.7, label='y') #alpha é a transparência
plt.hist(y2, alpha=0.7, label='y2')
plt.title('Histograma de y e y2')
plt.xlabel('nome do eixo x')
plt.ylabel('nome do eixo y')
plt.legend()
plt.savefig('histograma.png') # salva o gráfico
plt.show()
```

```
#boxplot
plt.boxplot([y, y2], labels=['y', 'y2'])
plt.ylabel('valor')
plt.title('Meus boxplots')
plt.show()
```
